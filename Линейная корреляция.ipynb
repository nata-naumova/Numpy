Коэффициент корреляции Пирсона - это мера линейной зависимости между двумя случайными величинами – X и Y.

```python
#Подключение библиотек
import pandas as pd #для анализа данных
import numpy as np #для обработки массивов
import matplotlib.pyplot as plt #для гистограмм, диаграмм
import seaborn as sns #для стат графиков
```

```python
df = pd.read_csv('data.csv', encoding='windows-1252') #чтение файла csv в объект dataframe
df.head() #возвращает записи из начала датафрейма
```

```python
corr = df[['price','bathrooms', 'sqft_living','sqft_above', 'sqft_basement','view']].corr()
corr.style.background_gradient(cmap='coolwarm')
print("Корреляционная матрица: \n",corr)
```

```python
sns.set(font_scale=1.15) #установка параметров
plt.figure(figsize=(10,8)) 
plt.title("Корреляционная матрица", fontsize=16) #заголовок
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.heatmap(corr, annot=True) #тепловая карта
plt.tight_layout() #чтобы заголовок вписался в область рисунка
plt.show() #показать
```

```python
def scatter_plot(df, var1, var2, label):
    sns.set(font_scale=1.5)
    plt.figure(figsize=(10,8))
    plt.title(label, fontsize=16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    sns.regplot(x=df[var1],y=df[var2],  fit_reg=False)
    plt.tight_layout()
    plt.show()
```

```python
scatter_plot(df, 'price', 'bedrooms', "График рассеяния price - bedrooms")
scatter_plot(df, 'price', 'sqft_living', "График рассеяния price - sqft_living")
scatter_plot(df, 'price', 'sqft_above', "График рассеяния price - sqft_above")
scatter_plot(df, 'price', 'sqft_basement', "График рассеяния price - sqft_basement")
scatter_plot(df, 'price', 'view', "График рассеяния price - view")
```
