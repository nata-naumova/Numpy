### Таблица с домами
* date - дата продажи дома
* price - цена
* bedrooms - спальни кол-во	
* bathrooms - ванные комнаты кол-во
* sqft_living - площадь дома
* sqft_lot - кв.м участка
* floors - этажи
* waterfront - набережная	
* view - кол-во просмотров	
* condition - состояние дома
* sqft_above - кв.м без подвала	
* sqft_basement - кв.м. подвала
* yr_built - год постройки
* yr_renovated - год ремонта
* street - улица	
* city - город
* statezip - индекс почтовый
* country - страна (США)
